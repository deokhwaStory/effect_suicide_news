### introduce

**베르테르 효과(Werther Effect)**

#### 국외 연구

- Phillips(1974): 유명인의 자살에 대한 언론보도인한 자살률 급증으로 처음 입증
    * 1948 - 1968 뉴욕타입즈 분석
    * 자살보도↑ 1면 보도↑ ⇒ 모방자살↑
- stack(2000) : 사건 보도 분량, 보도 기간 및 횟수, 보도 수단, 보도 매체 종류 등 영향
                연예인이나 유명인 등의 경우 일반인보도 14.3배 더 증가
       (2003) : 미디어의 자살 관련 뉴스가 영향 미침 입증
       (2005) : 언론의 자살 보도가 자살 수 증가와 연관있음 입증
    => 사건 보도 분량, 보도 기간 및 횟수, 보도 수단, 보도 매체 종류 등 자살 보도 외적 특성 뿐 아니라 자살 보도 방식 또한 모방 자살에 중요 요인
- Gould (2003) 
    * 자살 보도가 자세할 수 록 더욱 부정적 영향
    * 보도 기간과 보도량 및 매체의 인지도 등이 모방 자살에 유발 효과에 밀접함
    * 기사 제목이 자극적이고 다수의 매체를 통해 반복적 보도시 증가
- Joiner(2005) 
    * 자살 원인을 단정적으로 기술하거나 자살에 공감적 표현, 자살에 대한 이유를 설명하는 것 만드로도 자살의 정당성 부여될 수 있음
- Pirikis(2006) 
    * 광범위한 보도, 선정적 보도, 자살 방법을 상세히 기술한 보도일수록 증가
    * 시간 연관성 : 자살 보도 이후 처음 3일 내 모방 자살 효과 정점에 달하고 점차 감소해 약 2주 정도에 효과 소멸
- Mueller(2017) 
    - 미디어의 베르테르 효과에 입각해 미디어의 책임감있는 보도의 중요성 인정하나 더 중요한 것은 자살 관한 미디어를 둘러싼 사회적 역학 관계에 주목함으로 자살 암시가 미디어와 사회 맥락에 어떻게 작용하는지에 대한 폭넓은 시각 필요함 강조 


#### 국내 연구

1. 미디어의 자살보도 태도와 자살보도 노출량이 자살 생각에 미치는 영향

    - 김연종(2005) 
        * 주요 일간지의 자살이나 자살자에 대한 이해심을 표출하거나 공감 표현하는 등 암묵적 동조가 자살에 미치는 영향 큼 강조
    - 배준성, 허태균(2010)
        * 자살 보도가 자살에 대한 정당성을 부여하고 더 나아가 자살자의 행동을 긍적적으로 인식할 수 있게 함
    - 오지희, 김민정(2014) 
        * 미디어의 자극적 자살 보도에 노출된 청소년일수록 자살에 대한 생각 증가함
    
2. 자살보도 방식
    - 김병철(2007) 
        * 주요 일간지와 방송 3사에 보도된 자살 관련 기사의 언어적 표현 분석
            - 자살보도 방식이 자살에 미치는 부정적 영향 크므로 관련 언어적 표현 개선 필요함 주장
    -      (2010)  
        * 잠재적 자살자가 언론, 특히 자살 사건을 보도할 때 자살 동기에 대한 추측을 흥미 위주로 다루거나 자살 경위 또는 방법이 지나치게 구체적, 선정적일 때 수용자에게 자살 충동 유발할 가능성있음
    - 배준성, 허태균(2010) 
        * 보도 내용 중 선정적, 자극적 내용들을 보다 많이 기억했으며, 자살이 문제 해결의 방법(수단)이 될 수 있다고 자각하게 되면 모방자살에 대한 더욱 긍적적 생각과 태도 생성될 수 있음

#### 자살 보도 실태

- 김연종(2005) 
    * 2003년 12월 ~ 2005년 3월, 국내 4대 일간지 중심
    * 연예인 등 유명인의 자살 발생 후 언론사의 보도량 증가, 보도 방식에서는 자살 행위나 자살자에 대한 이해, 동정을 표하는 시각을 보인 기사가 약 11%, 자살자 미화가 21% 자살 방법(66.7%), 장소 공개(67.4%), 자살 이유를 당사자의 문제 극복을 위한 선택으로 보도한 기사가 10% 나타남
- 김병철(2007) 
    * 2006년 1년간 주요 일간지와 방송3사에 보도된 기사
    * '자살'이라는 어휘가 기사 제복에 자주 사용되고 있고, '투신','동반 자살', '한강' 등 자살 방법이나 장소를 상세히 언급한 어휘가 핵심어로 빈번히 사용됨
    * '살해', '경찰', '미스터리' 등의 용어를 사용해 자살을 범죄화하는 경우도 발견
- 박형민, 이민아(2009) 
    * 2008년 7월 ~ 2009년 6월 자상파 3사와 뉴스 전문 케이블 채널, 주요 일간지에서 보도된 자살 관련 기사 2,241건 분석
    * 연예인 자살 발생 후 선정적인 기사가 빈번히 나타났으며, 전체 자살 기사 중 37%가량의 기사는 선정적이고 자극적인 표현을 사용하여 작성됨
    * 기사 제목 : 자살 방법이 나타나는 경우 15%, 자살 장소 언급 31%, 자살 현장에 대한 구체적 묘사 14%
    * 기사 본문 : 자살 내용 기술 57%, 자살자의 성별, 연령, 직업 등 사적 정보 공개된 기사 55.2%, 자살원인으로 우울증을 보도한 경우 13.6%, 그외 악플, 가정불화, 이혼, 생활고, 사업 실패, 치정, 이성 문제 등의 자살 원인이 보도
        
- 김은이(2015) 
    * 2013년 1월 ~ 2014년 12월, 국내 신문 기사의 자살보도 내용 분석
    * 자살 기사의 자살 방법에 대한 정보가 기재된 경우 88.8%, 자살 장소에 대한 정보 94.9%, 자살 이유에 대한 정보 88.8%
